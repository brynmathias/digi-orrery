```python
%matplotlib inline
import matplotlib.pyplot as plt
from __future__ import print_function
import csv
raw_data = open("../test.txt")
coords = {}
for line in raw_data:
    vals = line.split(",")
    if vals[0] in coords:
        coords[vals[0]].append((float(vals[2]),float(vals[3]),float(vals[4]),float(vals[5]),int(vals[1]), vals[6], vals[7]))
    else:
        coords[vals[0]] = []
        coords[vals[0]].append((float(vals[2]),float(vals[3]),float(vals[4]),float(vals[5]),int(vals[1]), vals[6], vals[7]))

        earth_x_y = [[],[]]
for vec in coords['earth']:
    earth_x_y[0].append(vec[0])
    earth_x_y[1].append(vec[1])
    
moon_x_y = [[],[]]
for vec in coords['moon']:
    moon_x_y[0].append(vec[0])
    moon_x_y[1].append(vec[1])
    
sun_x_y = [[],[]]
for vec in coords['sun']:
    sun_x_y[0].append(vec[0])
    sun_x_y[1].append(vec[1])
```

```python
#,moon_x_y[0], moon_x_y[1], 'bo'
plt.plot(earth_x_y[0], earth_x_y[1], 'r--',  sun_x_y[0], sun_x_y[1], 'yo')
plt.show()
```
